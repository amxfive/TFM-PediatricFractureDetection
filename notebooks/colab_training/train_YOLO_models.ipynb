```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("banuprasadb/grazpedwri-dx")

print("Path to dataset files:", path)
```

```python
# Celda 4: Entrenamiento
from ultralytics import YOLO

# Cargamos el modelo Nano para que vuele
model = YOLO('yolov8n.pt')

# Entrenamos (notarás que ahora tarda 2-3 minutos por época)
results = model.train(
    data='/root/.cache/kagglehub/datasets/banuprasadb/grazpedwri-dx/versions/1/data.yaml',
    epochs=10,          # Ahora que tienes potencia, podemos subir a 25 épocas
    imgsz=640,
    device=0,           # Obligamos a usar la GPU (dispositivo 0)
    project='/content/', # Guarda los resultados en tu Drive
    name='graz_v1_colab'
)
```

```python
import shutil
# Crea un archivo .zip de toda la carpeta de resultados
shutil.make_archive('resultados_tfm', 'zip', '/content/graz_v1_colab')
```
